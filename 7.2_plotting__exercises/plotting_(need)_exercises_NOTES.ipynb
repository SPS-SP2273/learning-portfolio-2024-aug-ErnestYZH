- **plt.plot([x], y, [fmt], [x2], y2, [fmt2], ..., \*\*kwargs)** has many parameters
  - **x** and **y** represent the **x-values** and **y-values** of the data points
    - If an **even no.** of data points are provided, they will be plotted against each other in pairs
    - If an **odd no.** of data points are provided, the last set will be plotted against its length
  - **[fmt]**, or **format string** is a **string** that specifies formatting of the plot. It is an optional argument
    - Eg. **'ro'** for red circles
    - If not specified, it gives a line graph
  - **\*\*kwargs** are additional **keyword arguments** for customizing the plot
    - The order of these does not matter, and they are optional
    - **c=**, **ls=**, **lw=**, **marker=**, **label=**, etc
- All instances of **plt.plot()** before using **plt.show()** will create lines on the **same plot**

- **plt.errorbar()** works similarly to **plt.plot()**, but allows us to plot lines with **error bars**
  - There are 2 additional arguments: **xerr=** and **yerr=**, which gives the x-errors and y-errors respectively
    - The errors can be **lists** or **scalars**
      - Using a **list** will match each error to its corresponding point. The lists should be the same size
      - Using a **scalar** will set all of the errors in that axis to that value
    - These arguments are **optional** and **positional**
  - Unlike **plt.plot**, only 2 data sets (and hence 1 line) can be used with **plt.errorbar()**

- After **plt.plot()** and **plt.errorbar()**, which plots 1 (or more) lines each, we can use **additional functions** to specify more information of our plot
  - **plt.xlabel()** specifies the x-axis label
  - **plt.ylabel()** specifies the y-axis label
  - **plt.title()** specifies the plot's overall title
  - **plt.grid()** specifies the opacity of the grid
  - **plt.legend(loc=)** specifies the location of the legend
  - **plt.xlim()** specifies the upper and lower limits of the x-axis
  - **plt.ylim()** specifies the upper and lower limits of the y-axis
  - **plt.hlines(y, x1, x2)** plots a horizontal line
  - The above functions are optional and the plot will use default values if they are not called
  - The text-based functions have optional arguments such as **color=**, **fontsize=**, and **rotation=**

- We can use **plt.twinx()** to plot **2 different datasets** on the **same plot** with **different y-scales**
  - We separate the **plt.plot()** functions by **plt.twinx()**
  - The 2 lines will share the **same x-axis**
  - The first y-axis will appear on the left, while the second will appear on the right
  - We can call functions like **plt.legend()** and **plt.ylim()** separately for each instance of **plt.plot()**

- We can use **plt.style.use()** at the **start** to change the style of our plot
  - Use **plt.style.use('default')** to return to the default style

```python
from matplotlib import pyplot as plt
import numpy as np
# Trivial change for GitHub commit
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

**Exercise 2 (A simple mathematical plot)**

- Using **numpy arrays** for plots is often better since we can apply mathematical operations to them
- When we want to plot mathematical functions, we typically are not pluggin in datasets with points
  - We can use **x=np.linspace(start, stop, num=50)** to generate a numpy array of points to plot
  - We can then define **y** in terms of a mathematical function of **x**

```python
x = np.linspace(0, 50, 1000)
y = np.exp(-x/10) * np.sin(x)

plt.plot(x, y)
plt.xlabel('x',color='darkorange', fontsize=15)
plt.ylabel('y',color='darkorange', fontsize=15,rotation=0)
plt.grid(alpha=0.25)
plt.title(r'$y=e^{-x/10}\sin(x)$', fontsize=20)               # r string to ignore escape sequences
plt.ylim(-1, 1)
plt.show()
```
