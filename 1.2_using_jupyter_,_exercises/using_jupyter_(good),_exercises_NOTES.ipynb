**Exercise 1 (Learning to interrupt)**

Since I have set the condition in the while loop to **True**, the while loop will run without stopping.\
One way to stop it is to **interrupt** the **Kernal**

```python
while True:
    print('*', end='')
```

The **print()** function has the default argument **end='\n'**\
This means that every time the print function is called, it also prints **\n** after our string.\
In order to not have the newline, we have to specify the **keyword** as **end=''**.

```python
x = 0
while x < 10:
    print('*')
    x += 1
```

**Exercise 2 (Resetting & Restarting the Kernal)**

```python
print = 'PRINT'
```

```python
print('Hello World')
```

We have messed up the **print() function** by making **print** a **variable** and assigning it a **string**.\
If we try to run it now, it is 'replaced' with the string, which is not a function and can't be called.\
What is actually happening is demonstrated below:

```python
'PRINT'('Hello World')
```

By restarting the Kernal, all defined variables will be removed from the memory.\
This reinstates **print** as a function that can now be called.

**Exercise 3 (Timing Stuff)**

```python
%%timeit
squares = [i**2 for i in range(10)]
```

```python
print(squares)
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

```python
print(squares)
```

The numbers are referring to the time taken for each loop to execute in each example. Method 1 (using list comprehension) is faster than Method 2 (growing lists).

**%%timeit** runs the code in the entire cell 1M times, and calculates the **mean** and **std. dev.** of the time taken for each loop.\
List comprehension is faster as it calculates **i\*\*2** for each **i** before adding all to the list at once.\
Growing a list is slower as it **appends** each value of **i\*\*2** to the list before moving onto the next **i**, hence there are more steps.
